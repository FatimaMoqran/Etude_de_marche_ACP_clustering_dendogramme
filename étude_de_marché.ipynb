## CONSTRUCTION DU  DENDOGRAMME 

```python
def plot_dendrogram(Z, names):
    plt.figure(figsize=(10,25))
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('distance')
    dendrogram(
        Z,
        labels = names,
        orientation = "left"
    )
    plt.axvline(x=7.5, c='k')
    plt.show()

```

```python
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn import preprocessing

# préparation des données pour le clustering
X = dt_pca .values
names = dt_pca.index

# Centrage et Réduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

# Clustering hiérarchique
Z = linkage(X_scaled, 'ward')

# Affichage du dendrogramme
plot_dendrogram(Z, names)
```

```python
# DÉCOUPAGE DU DENDOGRAMME EN 5 GROUPES
plt.figure(figsize=(10,25))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('distance')
dendrogram(
        Z,
        labels = names,
        orientation = "left",
    color_threshold=7.5
    )
plt.axvline(x=7.5, c='k')

#plt.show()
plt.savefig('dendogramme')
```

```python
# DÉCOUPAGE DU DENDOGRAMME EN 5 GROUPES
plt.figure(figsize=(10,15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('distance')
dendrogram(
        Z,
        labels = names,
        orientation = "left",
    color_threshold=7.5
    )
plt.axvline(x=7.5, c='k')

#plt.show()
plt.savefig('dendogramme2')
```

puis coupez-le afin d'obtenir 5 groupes.

## test d'adequation

```python
#on va utiliser le test de Kolmogorov-Smirnov pour tester si les variables suivent une loi normale 
from scipy.stats import ks_2samp
import numpy
```

```python
dt.columns
```

On teste H0 F = F0 contre H1: F=! F0 ou F0 est la fonction de répartition d'une loi normale 


```python
# test de la variable Food supply 
#on calcule d'abord la moyenne et l'écart type de la variable et on les passe dans la fonction 
import random
#on met un random seed pour avoir les même résultats.
np.random.seed(1234)
ks_2samp(dt["Food supply (kcal/capita/day)"],list(numpy.random.normal(numpy.mean(dt['Food supply (kcal/capita/day)']), numpy.std(dt['Food supply (kcal/capita/day)']),1000)))
```

L'écart maximum entre la fonction de répartition de la variable et la gaussiène est de 8% et une p value de 26%
Au seuil de 5% on ne peut pas rejeter H0 l'hypothèse de normalité de la fonction de répartition de la varaiable Fodd supply ne peut pas être rejeté

```python
# test de la variable Protein supply quantity
np.random.seed(1234)
ks_2samp(dt["Protein supply quantity (g/capita/day)"],list(numpy.random.normal(numpy.mean(dt['Protein supply quantity (g/capita/day)']), numpy.std(dt['Protein supply quantity (g/capita/day)']), 1000)))
```

## test de comparaison

choisissez 2 clusters parmi ceux que vous aurez déterminé. Sur ces 2 clusters, testez la variable gaussienne grâce à un test de comparaison.

```python

cluster1 = groupe1['Food supply (kcal/capita/day)']
cluster3 = groupe3['Food supply (kcal/capita/day)']
```

```python
# on teste d'abord l'égalité des variances sur la variable food supply groupe 1 et groupe 4 
from scipy.stats import bartlett
bartlett(cluster1,cluster3)
```

on a une p value à 5%  ON peut pas  rejeter l'hypothèse nulle selon laquelle les deux clusters ont la même variance

```python
#On teste ensuite l’égalité des moyennes 
from scipy.stats import ttest_ind
ttest_ind(cluster1,cluster3)
```

on a une p value inferieure à 5% On peut rejeter l'hypothèse nulle selon laquelle les deux clusters ont la même moyenne. On peut en déduire que les deux clusters sont différents.
